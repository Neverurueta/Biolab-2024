#  **Ejercicio 1. Medidas de localización**



```python
T = np.array([3.4, 2.5, 4.8, 2.9, 3.6,
              2.8, 3.3, 5.6, 3.7, 2.8,
              4.4, 4.0, 5.2, 3.0, 4.8])
T
```

```python
# Medidas de localización
print('En promedio el tiempo de secado de la pintura esmaltada es:', T.mean().round(1),'horas',
      '\n El 50% de los tiempos de secados están por debajo de', np.median(T).round(1),'horas',
      '\n  El 75% de los tiempos de secados están por debajo de:', np.percentile(T, 75).round(1),'horas',
      '\n El 30% de los tiempos de secados están por debajo de:', np.percentile(T, 30).round(1),'horas')
```

#  **Ejercicio 2. Medidas de variabilidad**

```python
#Medidas de Variabilidad
print('La dispersión de los datos con respecto al promedio es de:', np.std(T).round(2),'horas',
      '\n lA varianza es:', np.var(T).round(2),'horas',
      '\n  El rango del conjunto de mediciones de tiempo de secado fue de:', (max(T)-min(T)).round(1),'horas',
      '\n Coeficiente de variación:', (np.std(T) / np.mean(T) * 100).round(0),'%')
```

```python
# Otra forma
import pandas as pd
pd.DataFrame(T, columns=['Tiempo de Secado']).describe()
```

# Otro Diagrama de caja y bigote - con dos poblaciones

```python
import seaborn as sns # Otra librería para gráficos
```

```python
propinas = sns.load_dataset('tips')
propinas.head()
```

```python
propinas.info()
```

```python
propinas.time.unique()
```

```python
with sns.axes_style(style='ticks'):
    g = sns.catplot(x='day', y='total_bill', hue='sex', data=propinas, kind='box')
    g.set_axis_labels("Day", "Total Bill");
```

```python
with sns.axes_style(style='ticks'):
    g = sns.catplot(x='day', y='total_bill', hue='time', data=propinas, kind='box')
    g.set_axis_labels("Day", "Total Bill");
```

# Probemos con un histograma

```python
propinas['tip_pct'] = 100 * propinas['tip'] / propinas['total_bill']

grid = sns.FacetGrid(propinas, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```
