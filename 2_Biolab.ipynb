## **Ejemplo. Distribución rectangular**

El tiempo de procesamiento en un equipo automatizado de una muestra
clínica a cierta temperatura y humedad ambiente según el laboratorio clínico
es de 2 horas, 5 minutos. En realidad, los tiempos de procesamientos de es
tipo de muestra están distribuidos uniformemente entre 2 horas y 2 horas,
20 minutos. Dado la informacióon anterior:

* Realice la gráfica de la función de densidad de probabilidad del
tiempo de procesamiento.




```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definimos el rabngo y el tamaño del mismo
valor_inferior = 120
valor_superior = 140
```

```python
x = np.linspace(valor_inferior, valor_superior, 100)  # 100 puntos para la gráfica
```

```python
# Calcular la altura de la distribución (probabilidad constante)
altura = 1 / (valor_superior - valor_inferior)
```

```python
plt.plot(x, np.full_like(x, altura), c='b')  # np.full_like crea un array con la misma forma que x, lleno de la altura
plt.xlabel('Valores')
plt.ylabel('Probabilidad')
plt.title('Distribución Rectangular')
plt.ylim(0, altura * 1.2)  # Ajustar el límite del eje y para una mejor visualización
plt.fill_between(x,0,altura, where= (x>=0) , color='orange', alpha=0.3)
plt.show()
```

* Cuál es la probabilidad de que una muestra no se retrase más de 5 minutos en su procesamiento?

P(125 < x < 130)

```python
n_limite_inferior = 125
n_limite_superior = 130
base = n_limite_superior - n_limite_inferior
probabilidad = altura * base
print('La probabilidad de que una musetra no se retrase más de 5 minuto es del', probabilidad*100,'%')
```

* ¿De que una muestra no se retrase máss de 10 minutos en su
procesamiento?

P(X > 135)

```python
n_limite_inferior = 135
base = valor_superior - n_limite_inferior
probabilidad = altura * base
print('La probabilidad de que una musetra se retrase más de 10 minuto es del', probabilidad*100,'%')
```

* ¿Cuál es el tiempo de procesamiento esperado?

```python
print('El tiempo de procesamiento esperado es de', (valor_superior + valor_inferior)/2, 'minutos')
```

* ¿Cuál es la variabilidad en el tiempo de procesamiento?

```python
print('La variabilidad en el tiempo de procesamiento es de', ((valor_superior - valor_inferior))/np.sqrt(12), 'minutos')
```

## **Ejemplo. Distribución triangular**

Durante el mes de agosto del año 2010, en Santiago de Compostela, las temperaturas mínima y máxima absolutas fueron de 12,2 ºC y 35,8ºC, respectivamente, y el valor más probable fue de 19,8ºC. Si se asume que la
temperatura sigue una distribución triangular de parámetros a=12,2, c=19,8
y b=35,8,.

1. ¿cu´al es la probabilidad de que supere los 30ºC?

```python
a = 12.2  # Valor mínimo
b = 35.8  # Valor máximo
c = 19.8   # Valor modal
```

```python
x = 30  # Límite inferior del intervalo
```

```python
if x<c:
  h = 2*(x-a)/((b-a)*(c-a))
elif x==c:
  h=2/(b-a)
elif x>c:
  h=2*(b-x)/((b-a)*(b-c))
```

```python
orientacion = input('digite la  desigualdad orientación del triangulo')
```

```python
if orientacion == '>' and x>c:
  base = b-x
  P = (base*h)/2
elif orientacion == '>' and x<c:
  base = x-a
  P_0 = (base*h)/2
  P = ((b-a)*(2/(b-a)))-P_0
elif orientacion == '<' and x<c:
  base = x-a
  P = (base*h)/2
else :
  base = b-c
  P_0 = (base*h)/2
  P = ((b-a)*(2/(b-a)))-P_0

print('La probabilidad de que la temperatura en Santiago de Compostela sea mayor de 30°C es del:', np.round(P*100,1),'%')
```

```python
# Otra forma
Acumulada = 1 -((b-x)**2/((b-a)*(b-c))) #Hallamos la función acumulada hasta x
Prob = 1- Acumulada # el área de la distribución es 1
print('La probabilidad de que la temperatura en Santiago de Compostela sea mayor de 30°C es del:', np.round(Prob*100,1),'%')
```

## **Ejemplo 1. Distribución Normal**

Dada una distribución normal estándar, calcule el área bajo la curva que se localiza:

a. A la derecha de z = 1.84

```python
from scipy.stats import norm
```

```python
mu_= 0  # Media de la distribución
sigma = 1  # Desviación estándar de la distribución
```

```python
z = 1.84  # Valor de la variable z
Acumulada = norm.cdf(z, loc=mu_, scale=sigma)
print(1-Acumulada)
```

b. Entre z = –1.97 y z = 0.86.

```python
z1=-1.97
z2=0.86
Acumulada1 = norm.cdf(z1, loc=mu_, scale=sigma)
Acumulada2 = norm.cdf(z2, loc=mu_, scale=sigma)
print(Acumulada2-Acumulada1)
```
